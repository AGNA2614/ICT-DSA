```python
import io
df=pd.read_csv(io.BytesIO(uploaded['Sales_add.csv']))
df
```

```python
# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing

# average of sales before digital add
print('Before digital add :',df['Sales_before_digital_add(in $)'].mean())

# average of sales after digital add
print('After digital add  :',df['Sales_After_digital_add(in $)'].mean())

# difference in sales
print('Difference in sales:',df['Sales_After_digital_add(in $)'].mean() - df['Sales_before_digital_add(in $)'].mean())
```
